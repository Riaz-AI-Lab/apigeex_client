```python
import requests
import os
import json
from dotenv import load_dotenv
load_dotenv()
key = os.getenv('APIGEE_X_KEY')
secret = os.getenv('APIGEE_X_SECRET')

def generate_token(key, secret):    
    # print(key)
    # print(secret)
    url = "https://dev.mcc.apix.mayo.edu/oauth/token"
    payload = f'grant_type=client_credentials&client_id={key}&client_secret={secret}'
    headers = {
      'Content-Type': 'application/x-www-form-urlencoded'
    }
    
    response = requests.request("POST", url, headers=headers, data=payload)    
    print("Token request code:", response.status_code)
    if response.status_code == 200:
        print("Token acquired successfully.")
    return response.json()['access_token']

token = generate_token(key, secret)

# Try LPR API
print("\n1. =============== TRY LPR API ===============")
response = requests.get(
    'https://dev.mcc.apix.mayo.edu/v3/mcc/fhir/r4/Patient?family:exact=Smith',
    headers = { 'Authorization': f'Bearer {token}' }
)

if response.status_code == 200:
    print("LPR request successfully")
    json_content = response.json()
    print(f"JSON Content: {json_content}")
else:
    # If not JSON, access as plain text
    print("ERROR with status:", response.status_code)
    text_content = response.text
    print(f"Text Content: {text_content}")


# Try LPR Metadata API
print("\n2. =============== TRY LPR METADATA API ===============")
response = requests.get(
    'https://dev.mcc.apix.mayo.edu/dna-lpr-fhir-metadata-internal',
    headers = { 'Authorization': f'Bearer {token}' }
)

if response.status_code == 200:
    print("LPR request successfully")
    json_content = response.json()
    print(f"JSON Content: {json_content}")
else:
    # If not JSON, access as plain text
    print("ERROR with status:", response.status_code)
    text_content = response.text
    print(f"Text Content: {text_content}")


# Try Clinical Trials API
print("\n3.=============== TRY CLINICAL TRIALS API ===============")
response = requests.get(
    'https://dev.mcc.apix.mayo.edu/dps-clinstudies/v1/first',
    headers = { 'Authorization': f'Bearer {token}' }
)

if response.status_code == 200:
    print("Clinical Trials request successfully")
    json_content = response.json()
    print(f"JSON Content: {json_content}")
else:
    # If not JSON, access as plain text
    print("ERROR with status:", response.status_code)
    text_content = response.text
    print(f"Text Content: {text_content}")
```
